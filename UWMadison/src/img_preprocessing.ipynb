```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
features = [
    "normal",
    "multiply",
    "anydepth"
]
fig = plt.figure(figsize=(18, 4*10))
ax = fig.subplots(10, 3)
plt_index = 0

def limitate(arr):
    return np.where(arr < 254, arr, 254)

for index, file_path in enumerate(glob.glob("../data/train/case2/case2_day3/**/*.png",  recursive=True)):
    if index < 50:
        continue
    data_npy = cv2.imread(file_path)
    if data_npy.sum():
        for feature in features:
            if feature == "normal":
                data_display = data_npy
                ax[plt_index][0].imshow(limitate(data_display))
            elif feature == "multiply":
                data_display = data_npy * 100
                ax[plt_index][1].imshow(limitate(data_display))
            elif feature == "anydepth":
                data_display = np.tile(np.expand_dims(cv2.imread(file_path, cv2.IMREAD_UNCHANGED), axis=-1), 3)
                ax[plt_index][2].imshow(limitate(data_display))
        plt_index += 1
            
            
    if plt_index == 10:
        break
plt.show()
```
